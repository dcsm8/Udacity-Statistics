```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
size = np.array([1700, 2100, 1900, 1300, 1600, 2200])
price = np.array([51000, 63000, 57000, 39000, 48000, 66000])

plt.figure()
plt.xlabel('Size')
plt.ylabel('House')
plt.scatter(size, price)
```

```python
fixed_price = price / size
fixed_price
```
